#### problema 11.Barbero
#### Barbero: Estilo de corte, silla asignada, productos de cuidado, clientes fijos, barbería y certificaciones de técnica.

```python
# --- SISTEMA DE IA: BARBER-TECH 2026 ---

# Definición de la lista principal que almacenará los registros
# Se inicializa vacía según tu solicitud
barberos = [] # [1]

# --- FUNCIONES DEL SISTEMA ---

def mostrar_menu():
    """Muestra las opciones disponibles en el sistema."""
    print("\n" + "="*60)
    print("      BARBER-BOT v5.0: ASISTENTE DE GESTIÓN PRO")
    print("="*60)
    print("1. Registrar nuevo Barbero (Entrada Key:Value)")
    print("2. Consultar Asistente de IA (Búsqueda por Técnica)")
    print("3. Ver Catálogo Completo (Tabla Alineada)")
    print("4. Apagar Sistema")
    print("="*60)
    print("*/*/*/*/*//*/*/*/*/*/*/*/*/*/*/*/**/*/*/")
    print("Programa realizado por Iris Chavez NL = 11")
def agregar_barbero():
    """Crea un diccionario individual y lo agrega a la lista 'barberos'."""
    # El diccionario empieza vacío antes de la entrada de datos [3]
    nuevo_barbero = {}

    print("\n--- REGISTRO DE NUEVO TALENTO ---")
    # Entrada de datos interactiva (Key, Values)
    nuevo_barbero["Estilo de corte"] = input("Estilo (clásico/moderno/urbano): ").lower()
    nuevo_barbero["silla asignada"] = input("Número de silla asignada: ")
    nuevo_barbero["productos de cuidado"] = input("Producto de cuidado recomendado: ").lower()
    nuevo_barbero["clientes fijos"] = input("Cantidad de clientes fijos: ")
    nuevo_barbero["barbería"] = input("Nombre de la barbería (sucursal): ")
    nuevo_barbero["certificaciones de técnica"] = input("Técnica (tijera/maquina/navaja): ").lower()

    # Se agrega el diccionario a la lista global 'barberos' [1]
    barberos.append(nuevo_barbero)
    print("\nIA: Registro guardado exitosamente en la lista 'barberos'.")

def asistente_ia():
    """Simula una IA analizando los diccionarios dentro de la lista."""
    print("\n--- ASISTENTE DE DECISIÓN IA ---")
    if not barberos: # Verifica si la lista está vacía [1]
        print("IA: La lista 'barberos' no tiene datos. Registre personal primero.")
    else:
        objetivo = input("IA: ¿Qué técnica busca el cliente hoy? (tijera/maquina/navaja): ").lower()
        encontrado = False

        # El ciclo for recorre la lista de diccionarios [5]
        for b in barberos:
            # Lógica de decisión comparando valores del diccionario [6]
            if b["certificaciones de técnica"] == objetivo:
                print(f"\n[SISTEMA]: Se recomienda al experto en estilo {b['Estilo de corte']}.")
                print(f"IA: Lo encontrará en la sucursal '{b['barbería']}', silla {b['silla asignada']}.")
                encontrado = True

        if not encontrado:
            print("IA: Lo siento, no contamos con expertos en esa técnica en este momento.")

def imprimir_tabla():
    """Muestra la lista de diccionarios en forma de tabla bien alineada."""
    # Encabezados con distribución fija usando f-strings para alineación [2]
    print("\n" + "-"*120)
    print(f"{'ESTILO':15} | {'SILLA':6} | {'PRODUCTO':18} | {'CLIENTES':10} | {'BARBERÍA':18} | {'TÉCNICA':15}")
    print("-" * 120)

    if not barberos:
        print(f"{'La lista de barberos está vacía actualmente.':^120}")
    else:
        # Recorrido de la lista para imprimir cada fila [5]
        for b in barberos:
            # Se alinean los valores de las llaves con el mismo ancho del encabezado [2]
            print(f"{b['Estilo de corte']:15} | {b['silla asignada']:6} | {b['productos de cuidado']:18} | "
                  f"{b['clientes fijos']:10} | {b['barbería']:18} | {b['certificaciones de técnica']:15}")
    print("-" * 120)

# --- CUERPO PRINCIPAL (Ciclo interactivo) ---
# Se utiliza While True para mantener el programa activo hasta que se decida salir [7]
while True:
    mostrar_menu()
    opcion = input("Seleccione una opción (1-4): ")

    if opcion == "1":
        agregar_barbero() # [8]
    elif opcion == "2":
        asistente_ia()
    elif opcion == "3":
        imprimir_tabla()
    elif opcion == "4":
        print("\nIA: Finalizando protocolos de gestión. ¡Hasta luego!")
        print("*/*/*/*/*//*/*/*/*/*/*/*/*/*/*/*/**/*/*/")
        print("Programa realizado por Iris Chavez NL = 11")
        break # Rompe el ciclo infinito para terminar el programa [7]
    else:
        print("\n[ERROR]: Opción no válida. Por favor, intente de nuevo.")

    print("*/*/*/*/*//*/*/*/*/*/*/*/*/*/*/*/**/*/*/")
    print("Programa realizado por Iris Chavez NL = 11")
```
